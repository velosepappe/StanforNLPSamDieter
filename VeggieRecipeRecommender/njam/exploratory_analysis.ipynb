```python
def listOccurrence(lst):
    occurrences = dict( (l, lst.count(l) ) for l in set(lst))
    return sorted(occurrences.items(), key=lambda x:x[1],reverse=True)
```

```python
#recepten met een 'veggie' label
recipeVeggie = []
for recipe in recipes:
    if not veggieLabels.isdisjoint(recipe['labels']):
            recipeVeggie.append(recipe)
```

```python
#alle componenten die in een veggie recept voorkomen
veggieComponents = []
for recipe in recipeVeggie:
        for ingredient in recipe['ingredients']:
            veggieComponents.append(ingredient['component'])
veggieComponentSet = set(veggieComponents)
veggieComponentsOccurrence = dict( (l, veggieComponents.count(l) ) for l in set(veggieComponents))
sorted(veggieComponentsOccurrence.items(), key=lambda x:x[1],reverse=True)
```

```python
#alle recepten zonder veggie label die enkel componenten bevatten die in 'veggie' recepten voorkomen
recipeProbablyVeggie = []
for recipe in recipes:
    if veggieLabels.isdisjoint(recipe['labels']):
        recipeComp = []
        for ingredient in recipe['ingredients']:
            recipeComp.append(ingredient['component'])
        if set(recipeComp).issubset(set(veggieComponents)):
            recipeProbablyVeggie.append(recipe)
len(recipeProbablyVeggie)
```

```python
#recepten zonder vlees label
recipeNonCarnie = []
for recipe in recipes:
    if carnieLabels.isdisjoint(recipe['labels']):
            recipeNonCarnie.append(recipe)
```

```python
#alle componenten die in een recept zonder vlees label voorkomen
nonCarnieComponents = []
for recipe in recipeNonCarnie:
        for ingredient in recipe['ingredients']:
            nonCarnieComponents.append(ingredient['component'])
nonCarnieComponentSet = set(nonCarnieComponents)
nonCarnieComponentsOccurrence = dict( (l, nonCarnieComponents.count(l) ) for l in set(nonCarnieComponents))
sorted(nonCarnieComponentsOccurrence.items(), key=lambda x:x[1],reverse=True)
```

```python
#recipes with a veggie label and a label for meats
recipeProbablyMiscategorisedVeggie = []
for recipe in recipes:
    if not veggieLabels.isdisjoint(recipe['labels']):
        if not carnieLabels.isdisjoint(recipe['labels']):
            recipeProbablyMiscategorisedVeggie.append(recipe)
recipeProbablyMiscategorisedVeggie
```

```python
#woorden of woordfragmenten die indiceren dat het recept een vegetarische variant is van een ander recept
veggieNameFragments = ['veggie','vegan','vegetarische']
veggieNames = []
for rec in recipes:
    if any(word in rec['name'].lower() for word in veggieNameFragments):
        veggieNames.append(rec['name'])
        
veggieNames
```

```python
[rec['name'] for rec in recipeVeggie]
```

```python
convertedNames = []
for name in veggieNames:
    for fragment in veggieNameFragments:
        if fragment in name.lower():
            convertedNames.append(name.lower().replace(fragment,'').rstrip().lstrip())
convertedNames
```

```python
#recepten die gedevegetariseerde recept namen bevatten. Potentiele items die geconverteerd kunnen worden
for rec in recipes:
    for nam in convertedNames:
        if nam.lower() in rec['name'].lower() and not any(word in rec['name'].lower() for word in veggieNameFragments):
            print(rec['name'])
```
