```python
import json
```

```python
with open('180508_data_njam.txt', encoding='utf8') as json_data:
    d = json.load(json_data)
    print(d)
```

```python
recipes = d['recipes']
len(recipes)
```

```python
def cleanSplitIngredients(ingredients):
    newIngredients = []
    for ingredient in ingredients:
        splits = ingredient['component'].split(" - ")
        newIngredient = {}
        for fragment in splits:
            newIngredient['component']= fragment
            newIngredient['amount']=ingredient['amount']
            newIngredient['unit']=ingredient['unit']
            newIngredients.append(newIngredient)
    return newIngredients
```

```python
operations = ['grofgesneden ', 'fijngehakte ','gehakte ','geraspte ','fijngesneden ','versgeraspte ','versgeperst ','fijngesnipperde ','vers gemalen ','gemalen ','versgemalen ',' van de molen','gekookte ','geroosterde ','hardgekookte ','gesmolten ','gepelde ','geklaarde ','kokend ','lauw ','koude ','zachte ',' (sap)', '(in blokjes)',' in blokjes',' in grove stukken',' in stukjes','opgeklopte ','geperst ','gebroken ','gesnipperde ']

def cleanOperationIngredients(ingredients):
    newIngredients = []
    for ingredient in ingredients:
        #some operations are separated by ", "
        splits = ingredient['component'].split(", ")
        newComponent = splits[0]
        for operation in operations:
            if operation in newComponent:
                newComponent = newComponent.replace(operation, '')
        newIngredient = {}
        newIngredient['component'] = newComponent
        newIngredient['amount'] = ingredient['amount']
        newIngredient['unit'] = ingredient['unit']
        newIngredients.append(newIngredient)
    return newIngredients
```

```python
units = ['bundeltje ','klein teentje ','dunne sneetjes ','dikke sneetjes ','sneetje ','handje ','handvol ','stronkje ', 'stronkjes ','blaadjes ','blaadje ','takje ','takjes ','bosje ','busseltje ','stengel ','stengels ','sprietjes ','teentje ','teentjes ','tenen ','plakjes ','druppels ','flinke scheut ','scheut ','scheutje ','flinke snuif ','snuif ','snuifje ','klontjes ', 'klontje ','portie ','partje ','cocktaillepel ','duimgroot stuk ','duimgroot stukje ','partje', 'stengels', 'blaadjes','scoop ','sap van 1 ','kneepje ','bolletje ','kopjes ','blik ','sneetjes ','handjes ']

def cleanUnitIngredients(ingredients):
    for ingredient in ingredients:
        for unit in units:
            if unit in ingredient['component']:
                print(unit + ' - ' + ingredient['component'])
                ingredient['component'] = ingredient['component'].replace(unit, '')
                ingredient['unit'] = unit.replace('^ ','').replace(' $','')
    return ingredients
```

```python
modifiers = ['rijpe ','grote ','middelgrote ','kleine ','verse ','vers ',' van de molen','zeste van ','stevige ']
def cleanModifierIngredients(ingredients):
    newIngredients = []
    for ingredient in ingredients:
        for modifier in modifiers:
            if modifier in ingredient['component']:
                ingredient['component'] = ingredient['component'].replace(modifier, '')
        ingredient['component'] = ingredient['component'].rstrip().lstrip()
    return ingredients
```

```python
def sortComponents(recipes):
    components = []
    for recipe in recipes:
        for ingredient in recipe['ingredients']:
            components.append(ingredient['component'])
    componentOccurrence = dict( (l, components.count(l) ) for l in set(components))
    return sorted(componentOccurrence.items(), key=lambda x:x[1],reverse=True)
```

```python
cleanedRecipes = []
for recipe in recipes:
    recipe['ingredients'] = cleanSplitIngredients(recipe['ingredients'])
    recipe['ingredients'] = cleanOperationIngredients(recipe['ingredients'])
    recipe['ingredients'] = cleanUnitIngredients(recipe['ingredients'])
    recipe['ingredients'] = cleanModifierIngredients(recipe['ingredients'])
    
```

```python
sorted = sortComponents(recipes)
print(len(sorted))
print(sorted)
```
