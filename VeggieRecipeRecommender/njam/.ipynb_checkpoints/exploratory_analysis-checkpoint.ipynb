```python
def listOccurrence(lst):
    occurrences = dict( (l, lst.count(l) ) for l in set(lst))
    return sorted(occurrences.items(), key=lambda x:x[1],reverse=True)
```

```python
def sortComponents(recipes):
    components = []
    for recipe in recipes:
        for ingredient in recipe['ingredients']:
            components.append(ingredient['component'])
    componentOccurrence = dict( (l, components.count(l) ) for l in set(components))
    return sorted(componentOccurrence.items(), key=lambda x:x[1],reverse=True)
```

```python
def componentDict(recipes):
    components = []
    for recipe in recipes:
        for ingredient in recipe['ingredients']:
            components.append(ingredient['component'])
    return dict( (l, components.count(l) ) for l in set(components))
```

```python
def searchComponent(name):
    matchingRecipes = []
    for rec in recipes:
        for ing in rec['ingredients']:
            if name in ing['component']:
                matchingRecipes.append(rec)
                break
    return matchingRecipes
```

```python
def suggestComponents(searchTerm, sortTerm):
    matchingRecipes = searchComponent(searchTerm)
    componentInMatchingRecipes = componentDict(matchingRecipes)
    A = len(matchingRecipes)
    print('Making suggestions for ' + searchTerm.upper())
    print('Matching recipes ' + str(len(matchingRecipes)) + '/' + str(len(recipes)))
    print('Coocurring components ' + str(len(componentInMatchingRecipes)) + '/' + str(len(referenceComponentDict)))
    print("('name',# as co-ingredient, # in total, sortterm")
    result = []
    for componentName, actualOccurrence in componentInMatchingRecipes.items():
        #avoid components that contain the searchterm
        if searchTerm not in componentName:
            sort = 1
            if sortTerm == 'odds' or sortTerm == 'oddslog':
                B = referenceComponentDict[componentName]
                notB = len(recipes) - B
                AandB = actualOccurrence
                AnotB = A-AandB
                BnotA = B-AandB
                notAnotB = len(recipes)-AandB-AnotB-BnotA
                sort = (AandB+1)*(notAnotB+1)/(AnotB+1)/(BnotA+1)
                if sortTerm == 'oddslog':
                    #general component get higher weight
                    sort = math.log(referenceComponentDict[componentName])*math.log(sort)
            elif sortTerm == 'log' or sortTerm == 'wtlog':
                occurrenceNonMatchingAddOne = referenceComponentDict[componentName] - actualOccurrence + 1
                actualAddOne = actualOccurrence + 1
                #expected occurrence
                expected = referenceComponentDict[componentName]*len(matchingRecipes)/len(recipes)
                #tbh, I don't know what this does anymore
                sort = math.log(actualAddOne*(len(recipes)-len(matchingRecipes))/occurrenceNonMatchingAddOne/len(matchingRecipes))
                if sortTerm == 'wtlog':
                    #general component get higher weight
                    sort = math.log(referenceComponentDict[componentName])*sort
            elif sortTerm == 'act/exp':
                expected = referenceComponentDict[componentName]*len(matchingRecipes)/len(recipes)
                sort = actualOccurrence/expected
            result.append({'comp':componentName, 'act':actualOccurrence, 'all':referenceComponentDict[componentName],sortTerm:sort})
    result.sort(key=lambda x: x[sortTerm], reverse=True)
    return result
```

```python
suggest = suggestComponents('zout','oddslog')
for item in suggest[:100]:
    print(item)
```

```python
def suggestComponentsWithSearchterm(searchTerm):
    matchingRecipes = searchComponent(searchTerm)
    componentInMatchingRecipes = componentDict(matchingRecipes)
    print('Making suggestions for ' + searchTerm.upper())
    print('Matching recipes ' + str(len(matchingRecipes)) + '/' + str(len(recipes)))
    print('Coocurring components ' + str(len(componentInMatchingRecipes)) + '/' + str(len(referenceComponentDict)))
    print("('name',weight, # as co-ingredient, # in total, # expected to occur, actual/expected occurrence)")
    result = []
    for componentName, actualOccurrence in componentInMatchingRecipes.items():
        occurrenceNonMatchingAddOne = referenceComponentDict[componentName] - actualOccurrence + 1
        actualAddOne = actualOccurrence + 1
        #expected occurrence
        expected = referenceComponentDict[componentName]*len(matchingRecipes)/len(recipes)
        result.append({'comp':componentName,'wt':actualAddOne/occurrenceNonMatchingAddOne, 'act':actualOccurrence, 'all':referenceComponentDict[componentName],'exp':"{0:.2f}".format(expected),'act/exp':"{0:.2f}".format(actualOccurrence/expected)})
    result.sort(key=lambda x: x['wt'], reverse=True)
    return result
```

```python
suggest = suggestComponentsWithSearchterm('kip')
for item in suggest:
    print(item)
```
